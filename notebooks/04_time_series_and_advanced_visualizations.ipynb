# Time Series and Advanced Visualizations

- Trends of launches and orbital types over years
- Comparative and multi-line charts


```python
import pandas as pd
import matplotlib.pyplot as plt

missions = pd.read_csv('../data/ISRO missions - till AUGUST 2022.csv')
missions.columns = missions.columns.str.replace('\xa0', ' ', regex=False).str.strip()
missions['Date'] = pd.to_datetime(missions['Date'], errors='coerce')
missions['year'] = missions['Date'].dt.year
```

```python
missions_per_year = missions['year'].value_counts().sort_index()

plt.figure(figsize=(10,5))
missions_per_year.plot(marker='o')
plt.title('Number of Missions Launched Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Missions')
plt.grid(True)
plt.show()
```

```python
orbital_year = missions.groupby(['year', 'orbital  type']).size().unstack(fill_value=0)
```

```python
missions.columns = missions.columns.str.replace('\xa0', ' ', regex=False).str.strip().str.replace(' +', ' ', regex=True)
```

```python
orbital_year = missions.groupby(['year', 'orbital type']).size().unstack(fill_value=0)
```

```python
orbital_year = missions.groupby(['year', 'orbital type']).size().unstack(fill_value=0)

plt.figure(figsize=(12,6))
orbital_year.plot()
plt.title('Missions by Orbital Type Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Missions')
plt.legend(title='Orbital Type')
plt.show()
```
