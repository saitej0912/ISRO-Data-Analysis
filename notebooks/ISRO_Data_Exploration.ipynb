```python
status_col = 'Status'  # Replace with actual column name

success_counts = missions[status_col].value_counts()
total_missions = success_counts.sum()
success_rate = success_counts.get('Success', 0) / total_missions * 100

print(f'Mission success rate: {success_rate:.2f}%')
```

```python
def mission_success(application):
    if isinstance(application, str) and 'partially successful' in application.lower():
        return 'Partial/Failure'
    else:
        return 'Success'

missions['mission_status'] = missions['Application'].apply(mission_success)

# Count success vs partial or failure
status_counts = missions['mission_status'].value_counts()

print(status_counts)

# Calculate success rate
success_rate = status_counts.get('Success', 0) / status_counts.sum() * 100
print(f"Estimated Mission Success Rate: {success_rate:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Satellites launched per year
satellites_per_year = satellites.groupby('year').size()

plt.figure(figsize=(10,5))
satellites_per_year.plot(kind='line', marker='o')
plt.title('Satellites Launched Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Satellites')
plt.grid(True)
plt.show()

# Missions launched per year
missions_per_year = missions.groupby('year').size()

plt.figure(figsize=(10,5))
missions_per_year.plot(kind='line', marker='o', color='green')
plt.title('Missions Launched Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Missions')
plt.grid(True)
plt.show()
```

```python
# Create pivot table of orbital types by year
orbital_yearly = missions.pivot_table(index='year', columns='orbital  type', aggfunc='size', fill_value=0)

plt.figure(figsize=(12,6))
for col in orbital_yearly.columns:
    plt.plot(orbital_yearly.index, orbital_yearly[col], marker='o', label=col)

plt.title('ISRO Missions by Orbital Type Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Missions')
plt.legend(title='Orbital Type')
plt.grid(True)
plt.show()
```
